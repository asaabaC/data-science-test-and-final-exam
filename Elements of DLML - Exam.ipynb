(c) What is the range of each quantitative predictor? You can answer this using the min() and max() methods

```python
import pandas as pd
numeric_columns = df.select_dtypes(include=['int64', 'float64'])


for column in numeric_columns.columns:
    predictor_range = df[column].max() - df[column].min()
    print(f"Range for {column}: {predictor_range}")
```

(d) What is the mean and standard deviation of each quantitative predictor

```python
import pandas as pd



numeric_columns = df.select_dtypes(include=['int64', 'float64'])

# Calculate the mean and standard deviation for each quantitative predictor
for column in numeric_columns.columns:
    predictor_mean = df[column].mean()
    predictor_std = df[column].std()
    print(f"Mean for {column}: {predictor_mean}")
    print(f"Standard Deviation for {column}: {predictor_std}\n")
```

(e) Now remove the 10th through 85th observations. What is the range, mean, and standard deviation of each predictor in the subset of the data that remains?

```python
import pandas as pd


df_subset = df.drop(df.index[9:84])


numeric_columns_subset = df_subset.select_dtypes(include=['int64', 'float64'])


for column in numeric_columns_subset.columns:
    predictor_range_subset = df_subset[column].max() - df_subset[column].min()
    predictor_mean_subset = df_subset[column].mean()
    predictor_std_subset = df_subset[column].std()
    
    print(f"Range for {column} in the subset: {predictor_range_subset}")
    print(f"Mean for {column} in the subset: {predictor_mean_subset}")
    print(f"Standard Deviation for {column} in the subset: {predictor_std_subset}\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


data = {'weight': [65, 70, 75, 80, 85, 90, 95],
        'age': [25, 30, 35, 40, 45, 50, 55]}

df = pd.DataFrame(data)


plt.hist(df['weight'], bins=10, edgecolor='black')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Distribution of Weight')
plt.show()


age = df['age']


pmf_age = np.histogram(age, bins=np.arange(min(age), max(age) + 1), density=True)
plt.bar(pmf_age[1][:-1], pmf_age[0], width=1.0, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('PMF')
plt.title('PMF of Age')
plt.show()
```
