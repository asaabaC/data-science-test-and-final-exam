## Question 2: Matrices and Numerical Python

Construct a matrix below ; 

$$\left[\begin{array}{ccc} 1 & 0.2 & 0.5 \\ 0.2 & 1 & 0.8 \\ 0.5 & 0.8 & 1 \end{array}\right]$$

as a NumPy array.


(a). Compute an Inversion of the matrix (`Transpose`) 

(b). Computer the (`determinant`) of the matrix if possible 

```python
import numpy as np

#firstly we create a matrix as a numpy array
matrix = np.array([[1, 0.2, 0.5],
                   [0.2, 1, 0.8],
                   [0.5, 0.8, 1]])
print(matrix)

# (a) Compute the inverse (transpose) of the matrix
inverse_matrix = np.linalg.inv(matrix)
print(inverse_matrix)

# (b) Compute the determinant of the matrix
determinant_matrix = np.linalg.det(matrix)

print(determinant_matrix)
```
